```python
import sys
import os
import numpy as np
import scipy.special

#LOCAL imports
from arcmodelling_IDMC import artificial_data_gen as adgen
from arcmodelling_IDMC import tools
from arcmodelling_IDMC import model_functions as mf
```



# 2. Add intensities and blur volume

## Default parameters 

```python
N=100
I = [26000., 31000., 42000.]
sigma_b=1.26

dataset_path='N'+str(N)
output_path='Ivol_01_noiseless_for_masking.npy'
```

```python
params={'I': I,
       'sigma_b': sigma_b}
```

## Load data

```python
centerdist_vol = np.load(dataset_path)
```

## Generate data

```python
#Sphere radius- "safety buffer" of 3 voxels
R = N/2 - 4*sigma_b - 3
```

```python
Ivol_noiseless = adgen.generate_Ivol(centerdist_vol, params, I1_ind=0, I2_ind=1, R=R)
```

```python
tools.plot_center_slices(Ivol_noiseless)
```

## Save data

```python
np.save(output_path, Ivol_noiseless)
```
